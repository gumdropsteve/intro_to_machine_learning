# KMeans Clustering

```
import pandas as pd

df = pd.read_csv('https://github.com/gumdropsteve/datasets/raw/master/iris.csv')

df
```

## What we know

```
import seaborn as sns

sns.pairplot(df, hue='species')
```

## What the algo know

```
import seaborn as sns

sns.pairplot(df)
```

https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html

```
from sklearn.cluster import KMeans
```

```
KMeans()
```

```
from sklearn.model_selection import train_test_split

X = df.drop(['species', 'target'], axis=1)
y = df.target

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.3)
```

```
km = KMeans()

km.fit(X_train)
```

```
km.labels_
```

```
km.predict(X_test)
```

```
km = KMeans()

km.fit(X)

km.labels_
```

```
from sklearn.metrics import accuracy_score

preds = km.labels_

accuracy_score(y, preds)
```

Doing it with 3 labels...

```
km = KMeans(n_clusters=3)

km.fit(X)

km.labels_
```

```
from sklearn.metrics import accuracy_score

preds = km.labels_

accuracy_score(y, preds)
```

```
y
```

Elbow Plot to Help Determine Amount of Clusters if you dont know how many exist


```
import matplotlib.pyplot as plt

inertia_ls = [] # save inertia values
for k in range(1, 11): # test different n_clusters between 1-10
    km = KMeans(n_clusters=k)
    km.fit(X) # apply KMeans
    inertia_ls.append(km.inertia_) # Append inertia to list

# Plot Elbow
plt.plot([i for i in range(1, 11)], inertia_ls, marker="o")
plt.title("Iris Elbow Plot")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia");
```

# Airlines

```
import pandas as pd

df = pd.read_parquet('https://github.com/gumdropsteve/datasets/raw/master/airlines.parquet')

df
```

```
corr = df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```
X = df[['CRSDepTime', 'CRSArrTime', 'ActualElapsedTime', 'Distance']]
y = df.ArrDelayBinary

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_tf = ss.fit_transform(X)
```

```
km = KMeans(n_clusters=2)

km.fit(X)

preds = km.labels_

accuracy_score(y, preds)
```

```
km = KMeans(n_clusters=2)

km.fit(X_tf)

preds = km.labels_

accuracy_score(y, preds)
```

```
import matplotlib.pyplot as plt

inertia_ls = [] # save inertia values
for k in range(1, 11): # test different n_clusters between 1-10
    km = KMeans(n_clusters=k)
    km.fit(X) # apply KMeans
    inertia_ls.append(km.inertia_) # Append inertia to list

# Plot Elbow
plt.plot([i for i in range(1, 11)], inertia_ls, marker="o")
plt.title("Airlines Elbow Plot")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia");
```
