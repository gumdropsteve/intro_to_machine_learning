## Baseline Model

```
def baseline_model(n_predictions, value_to_predict):
    """
    just predict a single value (e.g. mean) for everything
    """
    baseline_preds = []
    for i in range(n_predictions):
        baseline_preds.append(value_to_predict)
    return pd.Series(baseline_preds)
```

```
from sklearn.model_selection import train_test_split

X = df
y = df.price

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=4)
```

```
import numpy as np

n_predictions = len(y_test)  # how many predictions to make?
baseline_value = np.mean(y_train)  # what value to predict?

baseline_preds = baseline_model(n_predictions, baseline_value)
```

What are we predicting? Note: all predictions should be the same because this is a simple baseline model..

```
baseline_preds.tail(5)
```

```
from sklearn.metrics import mean_absolute_error

mean_absolute_error(y_test, baseline_preds)
```

```
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, baseline_preds)
```

## Avoiding Leakage
Improving the quality of the data...

```
df.info()
```

```
df.describe()
```

`url` can be used to identify a listing. As no 2 listings could share the same URL, each unique value in `url` represents 1 unique listing.. 

Source: [Data dictionary](https://github.com/gumdropsteve/intro_to_machine_learning/blob/main/day_10/data/README.md#data-dictionary)

```
len(df['url'].unique())
```

Because we have n (say 342) unique values in `url`, we have n (say 342) unique listings...

To avoid data leakage, we need to only keep 1 instance of each listing (`url`)... So let's drop the `ds` (datetime the info was pulled) and `search_filter` (search filter applied on top of the city filter) because the same listing (`url`) could show up multiple times under different `ds` and `search_filter` values...

Method: [pd.DataFrame.drop_duplicates() docs](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop_duplicates.html)

```
len(df.drop(['ds', 'search_filter'], axis=1).drop_duplicates())
```

That didn't cut it down to all the way... But if we add `subset='url'` then it should...

```
len(df.drop(['ds', 'search_filter'], axis=1).drop_duplicates(subset='url'))
```

By adding `subset='url'` we focus specifically on duplicate values in the `url` column..

```
dropped_df = df.drop(['ds', 'search_filter'], axis=1).drop_duplicates(subset='url')

dropped_df.sample(3)
```

So these should be the same..

```
len(dropped_df), len(dropped_df['url'].unique())
```

Now let's split the data and model once more..

```
X = dropped_df[['bedrooms', 'baths', 'shared_baths', 'avg_rating', 'n_reviews', 'gym_bool', 'pool_bool', 'dishwasher_bool']]
X = pd.concat([X, pd.get_dummies(dropped_df['type'])], axis=1)

y = dropped_df.price

# fill missing values in X with 0
X = X.fillna(0)

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=4)
```

## GridSearchCV
[LinearRegression Params](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

```
from sklearn.model_selection import GridSearchCV

lr = LinearRegression()

param_grid = {'fit_intercept':[True, False],
              'normalize': [False, True]}


grid = GridSearchCV(lr, param_grid)

grid.fit(X_train, y_train)

grid.best_params_
```

```
grid.best_estimator_
```

```
lr = grid.best_estimator_

lr.fit(X_train, y_train)
```

```
preds = lr.predict(X_test)

preds[-5: ]
```

```
r2_score(y_test, preds)
```

```
mean_absolute_error(y_test, preds)
```

```
mean_squared_error(y_test, preds)
```

[SGDR params](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDRegressor.html)

```
from sklearn.model_selection import GridSearchCV

sgdr = SGDRegressor()

param_grid = {'loss':['squared_loss', 'huber', 'epsilon_insensitive', 'squared_epsilon_insensitive'],
              'penalty': ['l1', 'l2', 'elasticnet'],
              'alpha': [0.0001, 0.001, 0.01, 0.1, 0.00001],
              'l1_ratio': [0.15, 0.10, 0.05, 0.30],
              'fit_intercept': [True, False],
              'max_iter': [1000, 2000, 3000, 4000, 5000, 6000, 7000],
              'shuffle': [True, False]}

grid = GridSearchCV(sgdr, param_grid)

grid.fit(X_train, y_train)

grid.best_params_
```

```
grid.best_estimator_
```

```
sgdr = grid.best_estimator_

sgdr.fit(X_train, y_train)
```

```
preds = sgdr.predict(X_test)

preds[-5: ]
```

```
r2_score(y_test, preds)
```
