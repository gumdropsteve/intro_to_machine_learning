## Data Wrangling
Improving the quality of the data...

`url` can be used to identify a listing. As no 2 listings could share the same URL, each unique value in `url` represents 1 unique listing.. 

Source: [Data dictionary](https://github.com/gumdropsteve/intro_to_machine_learning/blob/main/day_10/data/README.md#data-dictionary)

```
len(df['url'].unique())
```

Because we have 342 unique values in `url`, we have 342 unique listings...

To avoid data leakage, we need to only keep 1 instance of each listing (`url`)... So let's drop the `ds` (datetime the info was pulled) and `search_filter` (search filter applied on top of the city filter) because the same listing (`url`) could show up multiple times under different `ds` and `search_filter` values...

```
len(df.drop(['ds', 'search_filter'], axis=1).drop_duplicates())
```

That didn't cut it down to 342... But it's a start.

```
dropped_df = df.drop(['ds', 'search_filter'], axis=1).drop_duplicates()

dropped_df.sample(3)
```

```
len(dropped_df['url'].unique())
```

Let's add back the `ds` and `search_filter` columns and check out some of the `url` values that are still seen more than once..

```
df[['ds', 'search_filter']].iloc[dropped_df.index]
```

```
dropped_df[['ds', 'search_filter']] = df[['ds', 'search_filter']].iloc[dropped_df.index]

dropped_df.sample(3)
```

```
non_unique_urls = []

# go through the unique values in the url column
for _ in dropped_df.url.unique():
    # if this url is seen more than once
    if len(dropped_df.loc[dropped_df['url'] == _]) > 1:
        # add it to the non-unique urls list
        non_unique_urls.append(_)

len(non_unique_urls)
```

```
print(f'{len(non_unique_urls) / len(df.url.unique()) * 100}% of values in `url` are still not unique\nthis means those rows have the same value in `url` but different values in at least 1 column that is not `ds`, `search_filter` or `url`')
```

```
# what is the 0th non-unique URL
non_unique_urls[0]
```

```
# how many times is the 0th non-unique URL seen?
len(dropped_df.loc[dropped_df['url'] == non_unique_urls[0]])
```
